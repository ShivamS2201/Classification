# Task 1

```python
# formula for even fibonacci sequence En = 4En-1 + En-2, where En is the nth term in even fibonacci.

nter = int(input("Sum of Even Fibo Till What? ")) # Term till which sum is required.

#Initial Values
e1 = 0
e2 = 2 #Second value in even fibonacci (0,2,8,34...)

sums = e1 + e2 

while(e2<=nter):
    #print(e2,sums)
    e3 = 4 * e2 + e1 # Derived using recursion
    
    if e3 > nter: #stops the while if it reaches beyond the nter
        break
    
    e1 = e2
    e2 = e3
    
    sums = sums + e2
    
sums
```

```python
n = int(input("Enter N "))
# Using Sieve of Eratosthenes - (n log(logn))
    
prime = [True for i in range(n+1)]
p = 2
while (p * p <= n):
    # If prime[p] is not
    # changed, then it is a prime
    if (prime[p] == True):
        # Update all multiples of p
        for i in range(p * p, n+1, p):
            prime[i] = False
            p += 1
             
    # Print all prime numbers
for p in range(2, n+1):
    if prime[p]:
        print(p)
        
# Take a number, Computes its multiples greater then its square, then go to next unmarked number and repeat.
```

```python
# Tower of Hanoi with 3 Block and 3 Rods.

def TOH(n, source, destination, auxiliary):
    if n==1: 
        print ("Shift d- 1 from source->",source,"to destination->",destination)
        return
    
    TOH(n-1, source, auxiliary, destination) # recursive call to TOH func
    print ("Shift d-",n,"from source->",source,"to destination->",destination)
    TOH(n-1, auxiliary, destination, source)

# Blocks
n = 3
TOH(n,'A','B','C')
# A, C, B are rods

```

# Task 9

```python
"""
BUBBLE SORT -  sorting in which repeatedly swapping of the adjacent elements takes place unless the largest element in current
pass comes out. 
! Buuble sort goes till the end of the list/array and calls it a complete pass.
->Worst case : n^2 
->Best case: n 
"""
def bubble_sort(arr):
    arr_len = len(arr)
    
    for i in range(0,arr_len): # n
        for j in range(0,arr_len - i - 1): #  -1 is used to stop it from re-swapping the last element.
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j] # Swapping of variables ! Without 3rd var
                

array = [12,3,33,4,1,6,8,2]
bubble_sort(array)

array
```

### Binary Search

```python
def binarySearch (arr, l, r, x):
    
    # base case check
    if r >= l:
        mid = l + (r - l) // 2

    # If element is present at the middle.
        if arr[mid] == x:
            return mid
    
        elif arr[mid] > x:
            return binarySearch(arr, l, mid-1, x)

        else:
            return binarySearch(arr, mid + 1, r, x)

    else:
        return -1 # Acts as a flag.

arr = [ 2, 3, 4, 10, 40 ]
x = 10 # Element to be searched.

found = binarySearch(arr, 0, len(arr)-1, x)

if found != -1:
    print ("Element is at index % d" % found)
else:
    print ("Element is at in array")
```

### Linear Search

```python
def linearsearch(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

arr = ['i','a','o','u','e']
x = 'u'
print("Element at index "+str(linearsearch(arr,x)))
```
